```python
def NearestNeighbors(x_train,y_train,test_point,k=5):

    dist = euclidean_distances(test_point,x_train)
    nearest = zip(y_train, dist)
    nearest.sort(key=lambda x:x[1])
    nearest_k = nearest[:k]
    print nearest_k
    nearest_k = np.array(nearest_k)
    return nearest_k[:,0]
               
```

```python
def MostLikely(values):
    count = {}
    for val in values:
        if val not in count.keys(): count[val] = 1
        if val in count.keys(): count[val] += 1
    return max(count)
```

```python
#g implement a knn function

def knn(train_features,train_labels,test_features,test_labels=None,k=5):
    
    predictions=[]
    
    for point in test_features:
        predictions.append(MostLikely(NearestNeighbors(train_features,train_labels,point,k)))

    return predictions
```
